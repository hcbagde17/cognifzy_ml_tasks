```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('D:\cognifzy_ml_tasks\Dataset .csv')
df.head()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df['Restaurant ID'].nunique()
```

```python
df = df.drop(columns=['Locality Verbose','Restaurant ID'])
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Aggregate rating'], kde=True)
plt.title('Aggregate Rating Distribution')
plt.show()
```

```python
df.isnull().sum()
```

```python
columns = df.columns
num_cols = df.select_dtypes(include=['number']).columns.tolist()
cat_cols = df.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()
print("numeric columns:", num_cols)
print("categorical columns:", cat_cols)
```

```python
# visualizing discreate features against target variable
for feature in ['Country Code','Price range']:
    plt.figure(figsize=(7, 3))
    sns.barplot(x=feature, y='Aggregate rating', data=df, palette='viridis')
    plt.xticks(rotation=90)
    plt.title(f"{feature} vs Aggregate rating")
    plt.grid(True)
    plt.show()
```

```python
for feature in ['Average Cost for two', 'Votes']:
    if feature=="Aggregate rating":
        break
    else:
        plt.figure(figsize=(7, 3))
        sns.scatterplot(x=feature, y='Aggregate rating', data=df, color='teal')
        plt.title(f"{feature} vs Aggregate rating")
        plt.grid(True)
        plt.show()
```

```python
# visualizing categorical features against target variable
for feature in ['Currency', 'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu']:
    plt.figure(figsize=(7, 3))
    sns.barplot(x=feature, y='Aggregate rating', data=df, palette='viridis')
    plt.xticks(rotation=90)
    plt.title(f"{feature} vs Aggregate rating")
    plt.grid(True)
    plt.show()
```

```python
for col in cat_cols:
    print(f"Number of categories in {col}: {len(df[col].unique())}")
```
